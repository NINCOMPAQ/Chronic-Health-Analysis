Let's pull up our correlation matrix.

```python
import pandas as pd
df = pd.read_csv('processedCDC.csv', index_col='year_state')
correlations = df.corr()
#correlations
#uncomment the above line if you want to see the whole correlation matrix
```

```python
df
```

Now then, to query the correlates, type in queryCorrelates['Question, Demographic, valueType'].

```python
#Example, highest and lowest correlates for Coronary heart disease mortality, overall, crude rate
def queryCorrelates(qdv, howMany = 30, high=True, low=True):
    corVec = correlations[qdv]
    corVec = corVec.dropna(axis=0)
    corVec.sort_values(inplace=True)
    highest = corVec[:howMany]
    lowest = corVec[-howMany:]
    if(high):
        print(highest)
    if(low):
        print(lowest)
        

#Call the function here
QDV = 'Coronary heart disease mortality among all people, underlying cause, Overall, Crude Rate'
queryCorrelates(QDV)
```

The following function will find all question-demographic-valueType triplets correspond to a particular question and have (at least one point of) data available.

```python
def findColumns(question):
    cols = df.columns
    ourCols = cols[cols.str.startswith(question)]
    return ourCols

print(findColumns('Diseases of the heart mortality among all people, underlying cause'))
```

Then you can use those columns to query correlations, as per the following cell:

```python
queryCorrelates('Adults with any disability, White, non-Hispanic, Age-adjusted Prevalence')
```

Now, let's find a way to isolate just the questions with the topicID: CVD.

```python
CVDquestions = ('High blood pressure among adults',
       'Coronary heart disease mortality among all people, underlying cause',
       'Diseases of the heart mortality among all people, underlying cause',
       'High cholesterol among adults who have been screened',
       'Hospitalization for heart failure as principal diagnosis, Medicare-beneficiaries aged 65 years and older',
       'Taking medicine to control high blood pressure among adults with high blood pressure',
       'Cerebrovascular disease (stroke) mortality among all people, underlying cause',
       'Taking medicine for high cholesterol among adults')

cols = df.columns
CVDcols = cols[cols.str.startswith(CVDquestions)]
CVDcols
```

```python
aggregateCorrelates = correlations[CVDcols].sum(axis=1).sort_values(ascending=True)

aggregateCorrelates = aggregateCorrelates/224
aggregateCorrelates.to_csv('aggregateCorrelates.csv')
aggregateCorrelates
```

Let's get the aggregated correlates again, but only for the overall demographic

```python
overallcols = df.columns[df.columns.str.contains('Overall')]
overallcols
overalldf = df[overallcols]
overalldf
```

```python
overallCorr = overalldf.corr()
cols = overalldf.columns
CVDcols = cols[cols.str.startswith(CVDquestions)]
overallCorr[CVDcols]
```

```python
aggregateCorrelatesOverall = overallCorr[CVDcols].sum(axis=1).sort_values(ascending=True)
aggregateCorrelatesOverall = aggregateCorrelatesOverall/20
aggregateCorrelatesOverall.to_csv('aggregateCorrelatesOverall.csv')

aggregateCorrelatesOverall
```

```python
import pandas as pd
df = pd.read_csv('U.S._Chronic_Disease_Indicators.csv')
```

```python
#Getting our rows with year_state index

df.index = df['YearStart'].astype(str) + '_' + df['LocationAbbr']

states = df['LocationAbbr'].unique()
years = df['YearStart'].unique()
years.sort()
newdf = pd.DataFrame()
for state in states:
    placeholder = pd.DataFrame()
    placeholder['Year'] = years
    placeholder['State'] = state
    newdf = pd.concat([newdf,placeholder])
    
newdf.index = (newdf['Year'].astype(str) + '_' + newdf['State'])
newdf
```

```python
##The names of our columns will be in the following format:
##  question, demographic, valueType
questions = df['Question'].unique()
demographics = df['Stratification1'].unique()
vTypes = df['DataValueType'].unique()
cols = []
for question in questions:
    for demo in demographics:
        for type in vTypes:
            colName = question + ', ' + demo + ', ' + type
            cols.append(colName)
placeholder = pd.DataFrame(columns=cols)
newdf = pd.concat([newdf,placeholder])
```

```python
#Go through the original data row by row to get values
for index, row in df.iterrows():
    col = row['Question'] + ', ' + row['Stratification1'] + ', ' + row['DataValueType']
    newdf.loc[index, col] = row['DataValue']
    
newdf.dropna(axis=1, how='all', inplace=True)
newdf
```

```python
#Mean Value imputation and normalization
mathdf = newdf.drop(columns=['State'])
means = mathdf.mean(skipna=True)
processedDF = mathdf.fillna(means)
processedDF.to_csv('processedCDCnonnorm.csv')
processedDF = (processedDF-processedDF.min())/(processedDF.max()-processedDF.min())
processedDF.to_csv('processedCDC.csv', index_label='year_state')
processedDF
```

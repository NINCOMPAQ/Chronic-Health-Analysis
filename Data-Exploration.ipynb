```python
#This cell gets our (year, state, demographic) triplets for our rows

states = df['LocationAbbr'].unique()
years = list(df['YearStart'].unique())
demos = df['StratificationID1'].unique()
years.sort()
newdf = pd.DataFrame()
for state in states:
    placeholder = pd.DataFrame()
    placeholder['Year'] = years
    placeholder['State'] = state
    newdf = pd.concat([newdf,placeholder])
    
newdf.index = (newdf['Year'].astype(str) + '_' + newdf['State'])
newdf
    
```

```python
#This sets up our question columns. Right now they are all NaN

questions = df['Question'].unique()
for question in questions:
    sers = pd.Series()
    newdf = pd.concat([newdf,sers.rename(question)], axis = 1)
newdf = newdf.copy()
newdf
```

```python
#This function, when called, will fill in a specific question column with our data values
def grabNewCol(sers):
    question = sers.name
    slice = df[df['Question'] == question]
    for year in years:
        if question == 'Year' or question == 'State':
            break
        for state in states:
            thisIndex = str(year) + '_' + state
            print(thisIndex)
            print(question)
            thisDataValue = slice.loc[thisIndex, 'Question']
            print('safe')
            newdf.ix[thisIndex, question] = thisDataValue
            
#Now we call it on every column in our new data frame to fill in the values
newdf.apply(lambda x: grabNewCol(x))
print(newdf)
    
```
